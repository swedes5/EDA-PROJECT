```python
april_turnstiles_daily[["PREV_DATE", "PREV_ENTRIES"]] = (april_turnstiles_daily
                                                       .groupby(["C/A", "UNIT", "SCP", "STATION"])["DATE", "ENTRIES"]
                                                       .apply(lambda grp: grp.shift(1)))
```

```python
def get_daily_counts(row, max_counter):
    counter = row["ENTRIES"] - row["PREV_ENTRIES"]
    if counter < 0:
        # Maybe counter is reversed?
        counter = -counter
    if counter > max_counter:
        # Maybe counter was reset to 0? 
        print(row["ENTRIES"], row["PREV_ENTRIES"])
        counter = min(row["ENTRIES"], row["PREV_ENTRIES"])
    if counter > max_counter:
        # Check it again to make sure we're not still giving a counter that's too big
        return 0
    return counter
```

```python
april_turnstiles_daily["DAILY_ENTRIES"] = april_turnstiles_daily.apply(get_daily_counts, axis=1, max_counter=80000)
```

my goal is to help ensure that maintenance can be performed in an efficient manner so that commuters can feel minimal impact and that maintenance workers
can complete their jobs with as few distractions as possible.  One way to start this optimization is to check which days are typically less busy at different
stations throughout the MTA system.  For example, looking at three different stations, one being an extremely busy station, one being a station that recieves
an average number of commuters, and one being a less busy station, we can see if there are any patterns in ridership numbers and see if any given days of the 
week are best for doing general maintenance on the station or the turnstile level.  To further back this initial plan, we can look at a few general weeks.  One
set of data from the winter (first two weeks of the year) and another set of data from the summer (two weeks in july) to see if this proposed plan can work
